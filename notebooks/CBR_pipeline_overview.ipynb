# CBR Pipeline Overview

This notebook summarizes the 5-stage Case-Based Reasoning workflow used in this project.

## Pipeline Stages

1. Preprocessing PDF putusan into cleaned text files.
2. Case representation into CSV and JSON.
3. TF-IDF retrieval and query generation.
4. Case solution reuse with weighted similarity.
5. Evaluation of retrieval and prediction outputs.
